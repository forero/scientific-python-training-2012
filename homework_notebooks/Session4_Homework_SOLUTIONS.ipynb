## Part 1 - Write a Function

Using the methods we discussed in class, write a function that satisfies these conditions:

- Input is a filename of science image (FITS file), this can be any FITS file you like.

- Output is a PDF file with the image scaled reasonably for its content with a scaled colorbar representing the display range


Run the function to make sure it works and submit a copy of the PDF file produced.
Call it username_part1.pdf

```python
from pyraf import iraf
from iraf import noao, digiphot, daophot

import pyfits
import matplotlib.pyplot  as plt
import os
import numpy as np
```

```python
#file is a reserved word in many languages, it's best to use something else as your input variable
def display(filename):
    """Function to read in a fits file display it and create a pdf image"""
    image = pyfits.getdata(filename) #file here should be the text name of the file, with any extension information you might need
    plt.imshow(image,cmap=plt.cm.gray) #no scaling

def scaled_display(filename):
    """ make a reasonable guess at what the scaling should be """
    image=pyfits.getdata(filename)
    median=np.median(image) #most occuring value in the image, probably something good to center on, in an open field close to the sky value
    #you could use the median here and just scale above and below, or you could compute something more like the clipped median using masked arrays
    #to try and catch images which have problems in them and ignore those problems when computing your scale
    scale=median*3 #just as a guess, you could always use better logic here  or get more precise statistics, you probably have a well behaved dataset if the median is close to the mean
    pixels=np.where( np.abs(image) < scale, True, False) #return a bool array mapped to the image pixels where it's true when a pixel value is within 3 times the median
    median=np.median(image[pixels]) #refine the median
    std=np.std(image[pixels]) #refine the standard deviation
    plt.imshow(image,vmin=-3.*std, vmax=3.*std, cmap=plt.cm.gray) #show a better scaling
    
```

```python
"""
There is a little bit of confusion here in the difference between a command from the plt package and the figure object.

pyplot  keeps track of the current figure, so if you are only going to create one figure you dont't need to make the figure call 
if you are going to make several figures, then it's good practice to give them increasing numbers so you can reference them.
If you are going to make complex figures then it's easier to call the axes functions explicitly, you might want to use this method
in more complex functions

one way to use figure is like this:
"""
data=pyfits.getdata("../data/n8q624e8q_cal.fits") #from the class data directory
plt.clf() #clear any previous figure information if it exists
plt.figure(1) #now figure 1 is active, 1 is always the default figure and this isn't necessary
plt.imshow(data, vmin=-0.1, vmax = 40, cmap=plt.cm.gray)
plt.savefig('data.pdf')

plt.figure(2) #figure 2 active
plt.subplot(211) #make a subplot
plt.plot(range(100))
plt.title('figure 2 a')
plt.subplot(212) #make another subplot
plt.plot(range(200))
plt.title('figure 2 b')
plt.savefig('figure_2.pdf')
    

#There is also another way to deal with figure and plots, where you control the axes yourself. This is what you might want
#to try if you have several plots that share the same axis or space

fig=plt.figure(3)
fig.suptitle('MY TWO PLOTS')# here you can make a super title for the page
ax=fig.add_subplot(211) #2 down, 1 across, this is the top one
ax.plot(range(100)) #plot a line\a
ax.set_title('figure 3 a')
ax2=fig.add_subplot(212) #2 down, 1 across, this is the bottom one
ax2.plot(range(200)) #plot a different line
ax2.set_title('figure 3b')
fig.savefig('figure_3.pdf')
fig.tight_layout() #forces everything onto the page

#both of these methods, uses axes and not using axes, will produce the same results, the axes method just gives you
#some more fine grained control for fancy plots later and in scripting. I think we've shown examples of both of these during
#the sessions and perhaps this hasn't been as clear as it could be.


    
```

```python
%%file module.py
import os
import pyfits
import matplotlib.pyplot as plt

from pyraf import iraf
from iraf import noao,digiphot,daophot

def display(filename):
    """Function to read in a fits file display it and create a pdf image"""
    image = pyfits.getdata(filename) #file here should be the text name of the file, with any extension information you might need
    plt.imshow(np.log10(image),cmap=plt.cm.gray)
    
def find_objects(inputImage):
    """
    Find the objects in the field
    use DAOfind 
    """

    output_locations= inputImage  + ".stars" #best to set this if you're scripting
    
    #check if a file already exists and remove it
    if os.access(output_locations,os.F_OK):
        print "Removing previous star location file"
        os.remove(output_locations)
        
    
    #set up some finding parameters, you can make this more explicit
    iraf.daophot.findpars.threshold=3.0 #3sigma detections only
    iraf.daophot.findpars.nsigma=1.5 #width of convolution kernal in sigma
    iraf.daophot.findpars.ratio=1.0 #ratio of gaussian axes
    iraf.daophot.findpars.theta=0.
    iraf.daophot.findpars.sharplo=0.2 #lower bound on feature
    iraf.daophot.findpars.sharphi=1.0 #upper bound on feature
    iraf.daophot.findpars.roundlo=-1.0 #lower bound on roundness
    iraf.daophot.findpars.roundhi=1.0 #upper bound on roundness
    iraf.daophot.findpars.mkdetections="no"
    
    #assume the science extension
    sci="[SCI,1]"
    iraf.daofind(image=inputImage+sci,output=output_locations,interactive="no",verify="no",verbose="no")

    print "Saved output locations to %s"%(output_locations)
    
    return output_locations #return the name of the saved file
```

## A little bit about your environment and PATHs

When you make a file executable, meaning you'd like to run it from your shell prompt, you will be interacting with your shell PATHs.

Depending on your defaults you may, or may not, have the current working directory (./) on your environment PATH. If you try to execute your script
from the command line and you get back an error similar to "module.py not found", this is probably the cause. You either need to explicitly reference your script: "./module.py" or add "./" to your PATH so that it is automatically found.


To view what you have set as your PATH, important for running any executable on your machine
    > printenv PATH

To view what you ahve set as your PYTHONPATH, important for importing your script into other python scripts or a python session
    > printenv PYTHONPATH


The operating system will look on your PATHs and use the first directory which returns a match for what it's looking for. ORDER IS IMPORTANT. If you want to make sure you are running a certain version of something this is where you want to look. If you want to run a version of Python and open an ipython terminal with that version running for example, make sure that it is high up in your PATH statment. ipython is called from your shell command and the os will look on your PATH for it. 


You might not realize that you are setting up your environment PATHs anytime you use the STSci aliases for IRAFDEV and IRAFX. These just point to shell files which get run to setup your environment and point to the proper scripts for each of those packages:

    > which irafdev
    > irafdev:   aliased to set iraftype="dev !*"; source /usr/stsci/envconfig.mac/new_new_iraf_set_alias.csh
    

If you are creating a lot of your own scripts, modules, functions etc, it's good practice to store them in a common space. The location of this space should be on your PYTHONPATH so that the python interpreter can find them. 

Let's say I like to keep  my files (or modules) in ~sosey/python/scripts/, I would add that to my PYTHONPATH in the tcsh like this:

    > setenv PYTHONPATH ~/python/scripts/:${PYTHONPATH}

this statement puts the sosey directory in front of whatever else is already on the PYTHONPATH. If you do this a lot, you might want to add this to your .setenv or .cshrc to automatically execute when a new shell terminal is opened. If you only want to use it for a particular sesssion, do it manually. 

Now I can import anything in the ~/python/scripts/ directory into my python or pyraf session, or into whatever module I'm working on.
