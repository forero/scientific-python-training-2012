## Reading in Data and Basic Header Manipulations

## Lists

New Python lists are only created explicitly by calling

>>> list()

or making an empty list

>>> mylist = []

Executing a statement like

>>> a = b = []

will set a and b to the same list in memory.

Lists also have comprehension syntax that allows you to operate on each member in a compact way:

>>> new_list = [expression for variable in old_list]

```python
# This is same as selecting "*CORR" keywords above, but using Python list comprehension.
# That first thing in the list is what will be returned in the new list, in this case a tuple with 2 members (key, value)
mylist = [(key, hdr[key]) for key in header_keys if key.endswith('CORR')]
for key, val in mylist:
    print key, val
```

```python
# You can also convert lists to strings.
# You may use any delimiter you wish but make sure you can easily separate the items later.
# Using delimiter that is also in the data (e.g., ":" in data with RA in HH:MM:SS) is a bad idea.

# Join values in same row with equal sign, and different rows with comma.
mystring = ''
for mytuple in mylist:
    mystring += '='.join(mytuple) + ','
    
# Remove the last comma.
mystring = mystring[:-1]

print mystring
```

```python
# If you have a list of numbers (like ints) that you want to join in a string you can do the conversion at the same time.

# Make an array of 10 integers
numbers = np.arange(10)
print numbers

# For each item in the array, make it a string and then join the array together with comma
print ','.join(map(str, numbers))
```

```python
# Convert int to float (option 1)
float_numbers = map(float, numbers)
print type(float_numbers), float_numbers

# Convert int to float (option 2)
float_numbers = numbers.astype(np.float)
print type(float_numbers), float_numbers
```

```python
# Write the plotting commands as a function so we can use it again later.
def lec3_implot(image, vmin=-1.5, vmax=1.5, cmap=plt.cm.gray):
    fig, ax = plt.subplots()
    cax = ax.imshow(image['SCI'].data, vmin=vmin, vmax=vmax, alpha=1., cmap=cmap)
    ax.origin = 'lower'  # Set the image with (0,0) at lower left corner
    ax.set_title(image.filename() + '[SCI]')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    cbar = fig.colorbar(cax)  # Show colorbar scaling

    # We need this returned to overlay stuff later
    return ax
```

```python
# Display a quick image to see what we are working with.
ax = lec3_implot(image)
```

Now we have an output list of the stellar detections, lets read it in and overplot on our image to see how it did.

First, I'm going to do it by hand using regular python functions.  
Note: Parsing rules may vary for outputs from other tasks.

```python
with open(output, 'r') as stars:
    starpoints = stars.readlines()

# Store the header information in a different variable.
hlines = []
starloc = []
append_star = starloc.append  # You can assign a function to a variable!
for line in starpoints:
    if line.startswith('#'):
        # Strip off leading and trailing spaces and endline
        hlines.append(line.strip())
    else:
        # Each column becomes an element in a list.
        # Do not change to float yet in case there are INDEFs in there.
        # This is equivalent to starloc.append(...)
        append_star(line.split())
```

```python
# Extract column names.
# We know that our columns from daofind are labled in the 4th line
# from the bottom in the header section, and that they are the
# center of the line. 
# We do not care about the first and the last column paddings.
cols = hlines[-4].split()[1:-1]
print '***** Column names *****'
print 'Before:', hlines[-4]
print 'After:', cols
print

# Now we can reference our data array in nifty ways.
# Show (X, Y) location of the first star.
first_star = starloc[0]
idx_x = cols.index('XCENTER')
idx_y = cols.index('YCENTER')
print '***** First star info *****'
print 'Entire row:', first_star
print 'X Y:', first_star[idx_x], first_star[idx_y]
```

```python
# Save all the X and Y locations as floats.
# These are the actual catalog values in a useable form.

# Use list comprehension to make a list of (X,Y) tuples.
mytuple = [(float(line[idx_x]), float(line[idx_y])) for line in starloc]

# "Unzip" the list of tuples into two separate lists.
x, y = zip(*mytuple)

# Make sure the first elements belong to the first star above.
print 'First star:', x[0], y[0]

# Python will raise AssertionError if any of these tests fails.
# assert is useful in testing.
assert x[0] == float(first_star[idx_x])
assert y[0] == float(first_star[idx_y])
```

## Visualizing the Sources

```python
# Plot the image as above.
ax = lec3_implot(image)

# Overlay the stars we found as red points.
cax = ax.plot(x, y, 'r.', label='Star')
leg = ax.legend(numpoints=1)
```

```python
# There is an INDEF in one of the rows.
# Let's get rid of all rows with INDEF, and make a new clean starloc listing.
clean_stars = [line for line in starloc if 'INDEF' not in line]
print 'Number of rows thrown out:', len(starloc) - len(clean_stars)

# Extract sharpness and mag.
# Note that these do not correspond to x and y above because of the omitted rows.
# So we have to re-extract x and y too.
idx_sharp = cols.index('SHARPNESS')
idx_mag = cols.index('MAG')
mytuple = [(line[idx_x], line[idx_y], line[idx_sharp], line[idx_mag]) for line in clean_stars]
x, y, sharp, mag = zip(*mytuple)

# When dealing with numbers, numpy arrays work great.
# So we convert these to numpy float arrays.
x     = np.array(x, dtype=np.float)
y     = np.array(y, dtype=np.float)
sharp = np.array(sharp, dtype=np.float)
mag   = np.array(mag, dtype=np.float)

# Sanity check
print 'Second original row:', starloc[1]
print 'Second clean row:', clean_stars[1]
print 'Second clean X Y MAG SHARPNESS:', x[1], y[1], mag[1], sharp[1]
```

```python
# Plot of the sharpness versus the mag as blue pluses
fig, ax = plt.subplots()
cax = ax.plot(mag, sharp, 'b+')
ax.set_title('Sharpness vs. Mag from daofind')
ax.set_xlabel('Mag')
ax.set_ylabel('Sharpness')

# Save the plot as a PDF.
# Matplotlib automatically determines format from given extension.
plt.savefig('sharp_v_mag.pdf')
```

```python
# Select stars with sharpness greater than 0.6
idx_high = np.where(sharp > 0.6)

# Locations for those stars only
x_high = x[idx_high]
y_high = y[idx_high]
print 'Number of sharp stars:', x_high.size
```

```python
# Plot the image again in a different color.
ax = lec3_implot(image, vmin=-3., vmax=3., cmap=plt.cm.jet)

# Only show the high sharpness stars as black circles.
cax = ax.plot(x_high, y_high, 'ko', mfc='None')
```

## Analyze Photometry Results

```python
# Manipulate phot results using the astropy ascii package.
#
# As of 2012-11-28, there is a bug (Issue 509) and waiting for fix.
# This works if you update the codes with fix in Issue 510.
# The problem is that if your filename extends past the DAOPHOT format limit of 23 characters
# then the name of the file collides with the next column. That isn't an issue in this example because
# we have explicitly set the name of the stellar location and photometry file to be less than 23 characters
#
reader = ascii.Daophot()
photfile = reader.read(outphot)
```
